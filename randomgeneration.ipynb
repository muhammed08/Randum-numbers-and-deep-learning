```python
import random
import pandas as pd
```

```python
x_list = []
y_list = []
for i in range(2000):
    val=[]
    val.append(random.randint(0,20))
    val.append(random.randint(0,20))
    val.append(random.randint(0,20))
    x_list.append(val)
    val = random.randint(0,3)
    y_list.append(val)

```

```python
X = pd.DataFrame(x_list,columns = ["v1","v2","v3"])
X.head()
```

```python
Y = pd.DataFrame(y_list,columns = ["result"])
```

```python
from keras.utils.np_utils import to_categorical
```

```python
Y_cat = to_categorical(Y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y_cat,test_size = .2)
```

```python
from keras.models import Model, Sequential
from keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, MultiLabelBinarizer
from keras.layers.embeddings import Embedding
from keras.layers import Input
from keras.optimizers import Adam
```

```python
model1 = Sequential()
model1.add(Dense(10,input_shape=(3,)))
model1.add(Dense(4,activation = 'softmax'))
model1.compile (Adam(lr = .1), loss = 'categorical_crossentropy',metrics = ['accuracy'])
model1.summary()
```

```python
model1.fit(x_train,y_train,epochs = 100,validation_split = .1)
```

```python
loss, accuracy = model1.evaluate(x_test, y_test, verbose=0)
print('Accuracy: %f' % (accuracy*100))
```

```python
model2 = Sequential()
model2.add(Dense(100,input_shape=(3,)))
model2.add(Dense(1500))
model2.add(Dense(10))
model2.add(Dense(4,activation = 'softmax'))
model2.compile (Adam(lr = .1), loss = 'categorical_crossentropy',metrics = ['accuracy'])
model2.summary()
```

```python
model2.fit(x_train,y_train,epochs = 100,validation_split = .1)
```

```python
loss, accuracy = model2.evaluate(x_test, y_test, verbose=0)
print('Accuracy: %f' % (accuracy*100))
```
